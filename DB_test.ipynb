```python
import pymysql
```

```python
try:
    connect_db = pymysql.connect(host="localhost", user="root", password="", 
                                 database="online_book_store_db", charset='utf8' )
    print('連接成功')
except pymysql.Error as e:
    print('連接失敗' + str(e))
```

```python
with connect_db.cursor() as cursor:
    sql = """
    SELECT * from books
    """
    
    # 執行 SQL 指令
    cursor.execute(sql)
    
    # 取出全部資料
    data1 = cursor.fetchall()
    print(data1)

# 關閉 SQL 連線
# connect_db.close()
```

```python
with connect_db.cursor() as cursor:
    sql = """
    SELECT * from books
    JOIN categories ON books.category_id = categories.id
    JOIN cart ON books.id = cart.book_id
    """
    
    # 執行 SQL 指令
    cursor.execute(sql)
    
    # 取出全部資料
    data2 = cursor.fetchall()
    print(data2)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data2)
```
